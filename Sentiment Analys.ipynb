```python
from textblob import TextBlob

# Fungsi untuk mendapatkan sentimen
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df["sentiment"] = df["tweet"].apply(get_sentiment)

# Plot distribusi sentimen
sns.countplot(x=df["sentiment"], palette="coolwarm")
plt.title("Distribusi Sentimen")
plt.show()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download("vader_lexicon")

sia = SentimentIntensityAnalyzer()

df["emotion_score"] = df["tweet"].apply(lambda x: sia.polarity_scores(x)["compound"])

# Plot distribusi emosi
sns.histplot(df["emotion_score"], bins=20, kde=True, color="purple")
plt.title("Distribusi Skor Emosi dalam Tweet")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Buat Directed Graph (Graf Terarah)
G = nx.DiGraph()

# Tambahkan node dan edge berdasarkan retweet
for _, row in df.dropna(subset=["user_rt"]).iterrows():
    G.add_edge(row["user_rt"], row["username"])  # A -> B (B me-retweet A)

# Ambil 50 node dengan derajat tertinggi (paling banyak koneksi)
top_users = sorted(G.degree, key=lambda x: x[1], reverse=True)[:50]
top_nodes = [node[0] for node in top_users]
G_filtered = G.subgraph(top_nodes)

# Ukuran node berdasarkan jumlah koneksi (degree)
node_size = [G_filtered.degree(n) * 100 for n in G_filtered.nodes]

# Gunakan warna berdasarkan degree centrality (koneksi paling banyak)
centrality = nx.degree_centrality(G_filtered)
node_color = [centrality[n] for n in G_filtered.nodes]

# Plot graf dengan layout yang lebih jelas
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G_filtered, k=0.5)  # Layout lebih rapi
nx.draw(G_filtered, pos, with_labels=True, node_size=node_size, node_color=node_color, cmap="coolwarm", edge_color="gray", font_size=8)
plt.title("Social Network Analysis - Retweet Interaction", fontsize=14)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat Directed Graph
G = nx.DiGraph()

# Menambahkan edge berdasarkan retweet
for _, row in df.dropna(subset=["user_rt"]).iterrows():
    G.add_edge(row["user_rt"], row["username"])  # A -> B (B me-retweet A)

# Menghitung Degree Centrality
degree_centrality = nx.degree_centrality(G)

# Menampilkan Top 10 Akun dengan Degree Centrality Tertinggi
top_influencers = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

print("üîπ Top 10 Influencers (Degree Centrality):")
for user, centrality in top_influencers:
    print(f"{user}: {centrality:.4f}")

# Visualisasi Graf dengan Ukuran Node Berdasarkan Degree Centrality
plt.figure(figsize=(12, 8))
node_size = [v * 3000 for v in degree_centrality.values()]  # Perbesar ukuran node
nx.draw(G, with_labels=True, node_color="lightblue", edge_color="gray", node_size=node_size, font_size=8)
plt.title("Social Network Retweet Graph (Degree Centrality)")
plt.show()
```
